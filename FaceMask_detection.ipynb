Cleaning Dataset

```python
import os
multi = []

for dirpath, dirnames, filenames in os.walk("dataset"):

    for file in filenames:
        if file.endswith(".txt"):
            file_path = os.path.join(dirpath,file)
            with open(file_path,'r')as f:
                lines = f.readlines()
                if len(lines) > 1:
                    multi.append(file)

name_only_without_ext = [os.path.splitext(name)[0] for name in multi]

for name in name_only_without_ext:
    image_path = "dataset/"+name+".jpg"
    if os.path.isfile(image_path):
        os.remove(image_path)

for remove_file in multi:
    txt_path = "dataset" + '/' + remove_file
    if os.path.isfile(txt_path):
        os.remove(txt_path)
```

Splitting the data into train and validation

```python
import os
import cv2 as cv
import numpy as np

for dirpath,dirname ,filenames in  os.walk("dataset"):
    for file in filenames:
        if file.endswith('.txt'):
            txt_path = os.path.join(dirpath,file)

            image_path = os.path.join(dirpath,file.replace('.txt','.jpg'))
            image = cv.imread(image_path)
            height,width = image.shape[:2]
            r = max(width,height)/224


            with open(txt_path,'r') as t:

                lines = t.readlines()
                for line in lines:
                    box = line.split()
                    class_id = box[0]
                    x,y,w,h = map(float,box[1:5])

                    # x_pixel = x*width
                    # y_pixel = y*height
                    # w_pixel = w*width
                    # h_pixel = h*height

                    new_x = int((x-0.5*w)*w/r)
                    new_y = int((y-0.5*h)*h/r)
                    new_w = int(w*w/r)
                    new_h = int(h*h/r)

                    new_box = [class_id,new_x,new_y,new_w,new_h]

                    with open(txt_path,'w') as f:
                       f.write(" ".join(map(str, new_box)))



        elif file.endswith('.jpg'):
            image_path = os.path.join(dirpath,file)
            image = cv.imread(image_path)
            image = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
            height,width = image.shape[:2]
            r = max(width,height)/224
            new_width = int(width/r)
            new_height = int(height/r)
            resized_image = cv.resize(image,(new_width,new_height),interpolation=cv.INTER_AREA)
            black_canvas = np.zeros((224,224),dtype=np.uint8)
            black_canvas[0:new_height,0:new_width] = resized_image
            cv.imwrite(image_path,black_canvas)
```

```python
import os
os.mkdir("train")
os.mkdir("val")
os.mkdir("test")
```

```python
import os
import random
all_files = []
for dirpath, dirname, filenames in os.walk("dataset"):
    for file in filenames:
        if file.endswith('.jpg'):
            all_files.append(file)

random.shuffle(all_files)

train_len = int(len(all_files)*0.7)
val_len = int(len(all_files)*0.9)

train_data = all_files[:train_len]
val_data = all_files[train_len:]
test_data = all_files[val_len:]

for file in train_data:
    source_dir = os.path.join("dataset/",file)
    destination_dir = os.path.join("train/",file)
    os.rename(source_dir,destination_dir)

    file_txt = file.replace('.jpg','.txt')

    source_dir = os.path.join("dataset/",file_txt)
    destination_dir = os.path.join("train/",file_txt)
    os.rename(source_dir,destination_dir)

for file in val_data:
    source_dir = os.path.join("dataset/",file)
    destination_dir = os.path.join("val/",file)
    os.rename(source_dir,destination_dir)

    file_txt = file.replace('.jpg','.txt')

    source_dir = os.path.join("dataset/",file_txt)
    destination_dir = os.path.join("val/",file_txt)
    os.rename(source_dir,destination_dir)


for file in test_data:
    source_dir = os.path.join("dataset/",file)
    destination_dir = os.path.join("test/",file)
    os.rename(source_dir,destination_dir)

    file_txt = file.replace('.jpg','.txt')

    source_dir = os.path.join("dataset/",file_txt)
    destination_dir = os.path.join("test/",file_txt)
    os.rename(source_dir,destination_dir)
```
